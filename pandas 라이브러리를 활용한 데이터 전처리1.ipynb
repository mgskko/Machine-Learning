```python
# -DataFrame 행,열 삭제
df = pd.DataFrame(np.arange(12).reshape(3, 4),
                  columns=['A', 'B', 'C', 'D'])
```

```python
## ['B','C'] 컬럼을 삭제하겠다. axis=0 or 1 => 0:row 방향 , 1:col 방향
df.drop(['B', 'C'], axis=1)
```

```python
df.drop([0, 1], axis=0)
```

```python
# 데이터 수정
df.loc[1,'C'] = '육'
df
```

```python
df.iloc[1,2] = 6
```

```python
df
```

```python
#슬라이싱을 사용해 원하는 범위까지 바꾸기
df.iloc[:,1] = "끝"
df
```

```python
# 조건 필터링 가져오기
# 30대 이상만 가져오기
df[df['age']>=30]
```

```python
# job이 intern인 사람 가져오기
df[df['job'] == 'intern']
```

```python
# 조건이 여러개인 경우
# 30대 이상 40대 이하
df[(df['age']>=30) & (df['age']<=40)]
```

```python
# 30대 미만 혹은 40대 초과
df[(df['age']<30) | (df['age']>40)]
```

```python
5 in [1,3,5,8,23,215]
```

```python
# in을 통한 포함 조건 달기
df[df['job'].apply(lambda x : x in ['student','manager'])]
```
